## Merging

```python


df1 = pd.read_csv('Air Pollution.csv')
df1
```

```python
df2 = pd.read_csv('Weather.csv', encoding='euc-kr')
df2
```

```python
print(df1.columns)
print(df2.columns)
```

```python
# Rename columns

df1 = df1.rename(columns = {'날짜':'date'})
df1
```

```python
df1 = df1[['date','PM10','PM2.5']]
df1
```

```python
# Drop NaN values
df1=df1.dropna()
df1
```

```python
df1=df1.drop(0, axis=0)
df1
```

```python
df2 = df2.rename(columns={'일시': 'date', '풍속(m/s)': 'ws'})
df2
```

```python
df2 = df2[['date','ws']]
df2
```

## Merging

```python
# Checking data type
print(df1.dtypes)
print(df2.dtypes)
```

```python
df1 = df1.dropna().copy()
df2 = df2.dropna().copy()
```

```python
# To datetime
df1['date'] = pd.to_datetime(df1['date'], format='%m-%d-%H')

# Error because of the hours: 1-24 (should be 0-23)
```

```python
# One of the solutions
# Extracting last two characters from 'date' column because the data type is string.. 
temp = df1['date'].str[-2:]
temp = pd.to_numeric(temp)
temp
```

```python
temp = temp-1
temp2 = df1['date'].str[:-2] + temp.astype(str) + '-2021'
temp2
```

```python
df1['date'] = temp2
```

```python
df1['date'] = pd.to_datetime(df1['date'], format='%m-%d-%H-%Y')
df1
```

```python
# +1 hour

df1['date'] = df1['date'] + pd.DateOffset(hours=1)
df1

# Done!
```

```python
df1.dtypes
```

```python
# To datetime
df2['date'] = pd.to_datetime(df2['date'])
df2.dtypes
```

```python
# Merging!

df = pd.merge(df1, df2, how='inner', on='date')
df
```

```python
# How about outer?

pd.merge(df1, df2, how='outer', on='date')
```

```python
# Basic Indexing
df['PM2.5'] = pd.to_numeric(df['PM2.5'])
df.loc[df['PM2.5']>30]
```

```python
df.loc[df['PM2.5']==31]
```

```python
# And indexing
df['ws'] = pd.to_numeric(df['ws'])
df.loc[(df['PM2.5']>30) & (df['ws'] > 2.0)]
```

## Using function ('def')


```python
# Function can be applied in Pandas DataFrame

# Example: PM2.5 grade return
def grade(row):
    # conc. 0 - 15 -> Good
    # conc. 15 - 35 -> Normal
    # conc. 35 - 75 -> Bad
    # conc. over 75  -> Very bad
    if row['PM2.5'] > 75:
        return 'Very bad'
    elif row['PM2.5'] > 35:
        return 'Bad'
    elif row['PM2.5'] > 15:
        return 'Normal'
    else:
        return 'Good'
```

```python
df['Grade'] = df.apply(grade, axis=1)
df
```
